To find the value to minimize the cost function, there is a closed-form solution. In other words, a mathematical equation that gives the result directly. This is called the Normal Equation

$$\theta =  (X^T X)^{-1} X^T y$$

Let's generate some linear-looking data to test this equation

```python
import numpy as np
import matplotlib.pyplot as plt

X = 2 * np.random.rand(100,1)
y = 4 + 3 * X + np.random.randn(100,1)
plt.scatter(X,y,c='blue')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

Now, let's compute using the Normal Equation. We will use the inv() function from Numpy's linear algebra module (np.linalg) to compute the inverse of a matrix, and the dot() method for matrix multiplication:

```python
X_b = np.c_[np.ones((100,1)),X]
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
theta_best
```

Now, we can make predictions using $\theta$ :

```python
X_new = np.array([[0],[2]])
X_new_b = np.c_[np.ones((2,1)),X_new]
y_predict = X_new_b.dot(theta_best)
y_predict
```

Let's plot this model's predictions:

```python
plt.plot(X_new,y_predict,'r-')
plt.plot(X,y,'bo')
plt.xlabel('X')
plt.ylabel('y')
plt.axis([0, 2, 0, 15])
plt.legend(['Pediction'])
plt.show()
```

Performing Linear Regression using Scikit-Learn:

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X,y)
print('intercep:',lin_reg.intercept_)
print('coef:',lin_reg.coef_)
lin_reg.predict(X_new)
```

```python
plt.plot(X_new, lin_reg.predict(X_new), 'r-')
plt.plot(X, y, 'bo')
plt.legend(['Prediction'])
plt.show()
```
