## Logistic Regression

Some regression algorithms can be used for classification. Logistic Regression (also called Logit Regression) is commonly used to estimate the probability that an instance belongs to a particular class. If the estimated probability is greater than 50%, the the model predicts that the instance belong to that class, called the positive class, labeled "1", and otherwise it predicts that it does not, belong to negative class, labeled "0".

This makes it a "binary classfier"

Let's use iris dataset to simulate Logistic Regression

#### Iris dataset

Include length and width of petal and sepal, with 150 flowers belong to three different types: Setosa, Versicolour and Virginica

Let's try to build a classifier to detect Virginica iris base on width of petal

#### Downloading the data

```python
from sklearn import datasets
iris = datasets.load_iris() #dictionary
iris.keys()
```

Now, we will train Logistic Regression model

```python
X = iris['data'][:, 3:] #width petal
y = (iris['target']==2).astype(int) #class Virginica
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X,y)
```

```python
import numpy as np
X_test = np.linspace(0,3,1000).reshape(-1,1)
y_test = log_reg.predict_proba(X_test)
```

#### Visulazation

```python
import matplotlib.pyplot as plt
plt.plot(X,y,'r^')
plt.xlabel('Petal witdh')
plt.ylabel('Probability')
plt.show()
```

Petal width of Virginica has range 1.4cm to 2.5cm

```python
plt.plot(X,(iris['target']==0).astype(int),'ks')
plt.plot(X,(iris['target']==1).astype(int),'ks')
plt.xlabel('Petal witdh')
plt.ylabel('Probability')
plt.show()
```

Others iris has range 0.1cm to 1.8cm

Combine with predict curve

```python
plt.plot(X,y,'r^')
plt.plot(X,(iris['target']==0).astype(int),'ks')
plt.plot(X,(iris['target']==1).astype(int),'ks')
plt.plot(X_test, y_test[:,0],'b--',label='Not Iris Virginica')
plt.plot(X_test, y_test[:,1],'g-',label='Iris Virginica')
plt.xlabel('Petal witdh (cm)')
plt.ylabel('Probability')
plt.legend()
plt.show()
```

Generally, petal width over 2cm, model predicts is Virginica, because it has high probability.
And, petal width below 1cm, high probability it's not Virginica. 
Between 1cm and 2cm, it's incertainly.
If we use predict() method, it'll return class with the highest probability.
Therefore, there is a Decision Boundary around 1.6cm where both of probability is 50%. If petal width
over 1.6cm, it's Virginica. If petal width below 1.6cm, it's not Virginica.
