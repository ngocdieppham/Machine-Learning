```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
X,y = make_blobs(n_samples=100,n_features=2,centers=3, cluster_std=0.4, random_state=0)
#print(X)
#print(y)
kmean =KMeans(n_clusters=3)
kmean.fit(X)
clusters = kmean.predict(X)
print(clusters)
```

```python
print(*X.T) 
```

```python
plt.scatter(*X.T, c=clusters)
# c là array or list of color ==> use để show mỗi cluster theo color khác nhau
# usually, vd có 100 points là c phải có 100 colors

#cmap-colormap, bảng màu có sẵn, mỗi color có value từ 0-100
#defaults: cmap='viridis'
```
